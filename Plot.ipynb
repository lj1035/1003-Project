```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('median absolute error.csv', header=0, index_col=0)
df.head()
```

```python
sns.set_style("whitegrid")

# Set up the matplotlib figure
fig, ax = plt.subplots()

sns.barplot(df.index, df['Mean'], palette="GnBu")
ax.set_title('Median Absolute Error')
ax.set_ylabel("Model")
ax.set_xticklabels(df.index, rotation=45)

# Finalize the plot
sns.despine(bottom=True, left=True)
plt.show()
```

```python
df_plot2 = pd.read_csv('time performance.csv', header=0, index_col=0)
```

```python
sns.set_style("whitegrid")

# Set up the matplotlib figure
fig, ax = plt.subplots()
fig.set_size_inches(12, 4)

sns.barplot(df_plot2['Time'][::-1], df_plot2.index[::-1], palette="GnBu")
ax.set_title('Time Performance')
ax.set_xlabel('Time')

# Finalize the plot
sns.despine(bottom=True, left=True)
plt.show()
```
